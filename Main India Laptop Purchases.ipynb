```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.pie(data['Purchase_Mode'].value_counts(), labels=('Online', 'Offline'), autopct='%1.1f%%', wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white'})
plt.title('Online Vs Offline Purchases Among Customers')
plt.show()
```

```python
plt.pie(data['Income_Level'].value_counts(), labels=('Medium','Low','High'), autopct='%1.1f%%', wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white'})
plt.title('Income Level Distrubution Among Customers')
plt.show()
```

```python
# Medium Income Level customers make up half of the customer base.
```

```python
medium_income_data = data[data['Income_Level']=='Medium']
medium_income_data.head()
```

```python
sns.countplot(x=medium_income_data['City'], width=0.5, color='blue')
plt.title('Medium Income Level Customers By City', fontsize=18)
plt.xticks(rotation=45, ha='right')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

```python
# The distribution seems to be very similar between cities in terms of the amount of medium income level customers
```

```python
plt.pie(medium_income_data['Purchase_Mode'].value_counts(), labels=('Online', 'Offline'), autopct='%1.1f%%', wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white'})
plt.title('Offline vs Online Purchases of Medium Income Customers')
plt.show()
```

```python
high_income_data = data[data['Income_Level']=='High']
high_income_data.head()
```

```python
color_palette = {'Delhi': 'green', 'Jaipur': 'green', 'Mumbai': 'green', 'Chennai': 'green', 'Ahmedabad': 'green', 'Bangalore': 'red', 'Pune': 'green', 'Hyderabad': 'red', 'Kolkata': 'red'}
sns.countplot(x=high_income_data['City'], width=0.5, palette=color_palette)
plt.title('High Income Level Customers By City', fontsize=18)
plt.xticks(rotation=45, ha='right')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

```python
plt.pie(high_income_data['Purchase_Mode'].value_counts(), labels=('Online', 'Offline'), autopct='%1.1f%%', wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white'})
plt.title('Offline vs Online Purchases of High Income Customers')
plt.show()
```

```python
low_income_data = data[data['Income_Level']=='Low']
low_income_data.head()
```

```python
sns.countplot(x=low_income_data['City'], width=0.5, color='blue')
plt.title('Low Income Level Customers By City', fontsize=18)
plt.xticks(rotation=45, ha='right')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

```python
plt.pie(low_income_data['Purchase_Mode'].value_counts(), labels=('Online', 'Offline'), autopct='%1.1f%%', wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white'})
plt.title('Offline vs Online Purchases of Low Income Customers')
plt.show()
```

```python
data['Age'].describe()
```

```python
bins = [18, 24, 34, 44, 54, 60]
labels = ['18-24', '25-34', '35-44', '45-54', '55-60']
data['Age_Group'] = pd.cut(data['Age'], bins=bins, labels=labels)
data['Age_Group'].value_counts()
```

```python
plt.pie(data['Age_Group'].value_counts(), labels=('45-54', '35-44', '25-34', '18-24', '55-60'), autopct='%1.1f%%', wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white'})
plt.title('Age Group of Customers')
plt.show()
```
